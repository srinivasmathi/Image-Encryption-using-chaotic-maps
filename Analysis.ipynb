```python
import numpy as np
import matplotlib.pyplot as plt
X=[] #to store r values
Y=[] #to store logistic map values
R=np.linspace(2,4,10000) #will generate 10000 numbers between 2 and 4
for r in R:
  X.append(r)
  x=np.random.random() # initial x value
  for i in range(300):
    x=r*x*(1-x)
  Y.append(x)
plt.title("BIFURCATION DIAGRAM")
plt.xlabel("r values")
plt.ylabel("logistic map values")
plt.plot(X,Y,ls="",marker=",")
plt.show()
```

### Lyapunov Exponent


```python
import numpy as np
import matplotlib.pyplot as plt
R=np.linspace(2,4,20000)
X=[]
Y=[]

for r in R:
  X=[]
  x=np.random.random()
  for j in range(200):
    x=r*x*(1-x)
    X.append(np.log(abs(r-2*r*x)))
  Y.append(np.mean(X))
plt.title("LYAPUNOV EXPONENT")
plt.xlabel("r values")
plt.ylabel("Lyapunov value")
plt.plot(R,Y,marker=",",ls="")
plt.show()
```
